```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
import seaborn as sns

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) ## Adjusts the configuration of the plots 
```

```python
#Loading the Dataset

data = pd.read_csv(r'C:\Users\Mis\OneDrive\Desktop\datasets.csv')
```

```python
#Data Cleaning

data.isnull().sum()
```

```python
#Handling Null Values

#Hnadling Non-numeric Fields

data['neighbourhood'].fillna(data['neighbourhood'].mode()[0], inplace=True)
data['room_type'].fillna(data['room_type'].mode()[0], inplace=True)
data['last_review'].fillna(data['last_review'].mode()[0], inplace=True)

#Handling Numeric Fields

data['price'].fillna(data['price'].median(), inplace=True)
data['minimum_nights'].fillna(data['minimum_nights'].median(), inplace=True)
data['number_of_reviews'].fillna(data['number_of_reviews'].median(), inplace=True)
data['reviews_per_month'].fillna(data['reviews_per_month'].median(), inplace=True)
data['calculated_host_listings_count'].fillna(data['calculated_host_listings_count'].median(), inplace=True)
data['number_of_reviews_ltm'].fillna(data['number_of_reviews_ltm'].median(), inplace=True)
data['availability_365'].fillna(data['availability_365'].median()[0], inplace=True)

```

```python
# Dropping the records with Null Latitude or Longitude Values

data = data.dropna(subset=['latitude', 'longitude'])
```

```python
#Checking for Duplicate Records

data.duplicated().sum()

#Dropping Duplicate Values

#data[data.duplicated()]
data.drop_duplicates(inplace=True)
```

```python
data[data.duplicated()]
```

```python
#Type Casting

data['id'] = data['id'].astype(object)
data['host_id'] = data['host_id'].astype(object)
```

```python
#EDA
#Data Analysis
#Univarient Analysis

#Identifying Outliers in Price

df = data[data['price'] < 1500]
sns.boxplot(data=df, x='price')

plt.show()
```

```python
#Price Distribution

sns.histplot(data=df, x='price', bins=100)
plt.title("Price Distribution")
plt.ylabel('Frequency')

plt.show()
```

```python
#Availability Distribution

sns.histplot(data=df, x='availability_365')
plt.title('Availability Distribution')
plt.show()
```

```python
data.dtypes
```

```python
#Feature Engineering

df['price_per_bed'] = df['price']/df['beds']
df.head()
```

```python
#Checking mean bed price per unique neighbourhood

df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

```python
#Bi-varient Analysis

#Price Dependency on Neighbourhood

sns.barplot(data=df, x='neighbourhood_group', y='price_per_bed', hue='room_type')
plt.title('Price v/s Neighbourhood')
plt.ylabel('Price per Bed')
plt.xlabel('Neighbourhood')
plt.show()
```

```python
#Number of Reviews Dependency on Price

sns.scatterplot(data=df, x='number_of_reviews', y='price_per_bed', hue='neighbourhood_group')
plt.title('Price v/s Number of Reviews')
plt.ylabel('Price per Bed')
plt.xlabel('Number of Reviews')
plt.show()
```

```python
#Correlation for Numeric Features

correlation_matrix = df.corr()

#Visualizing the Correlation Matrix

sns.heatmap(correlation_matrix, annot=True)
```

```python
#Correlation for All Features

correlation_matrix = df_num.corr()

#Visualizing the Correlation Mattrix

sns.heatmap(correlation_matrix, annot=True)
```
